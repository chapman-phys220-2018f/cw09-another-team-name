## Introduction

There are many ways to understand and write a derivative. Specifically, the derivative of a function fff at a point xxx is defined as the limit:
$$f'(x) = \frac{df}{dx}(x) = Df(x) \equiv \lim_{h \to 0} \frac{f(x+h) - f(x)}{(x + h) - x} = \lim_{h\to 0} \frac{f(x) - f(x - h)}{x - (x - h)} = \lim_{h \to 0} \frac{f(x+h) - f(x-h)}{(x+h)-(x-h)}$$
The first limit definition is called the "forward difference", the second limit definition is called the "backward difference", and the third limit is called the "central difference". When there is enough information, it is preferred to use the central difference, due to its better accuracy. This finite difference operator can be represented by the matrix:
$$D=\begin{pmatrix}
    -2 & 2 & 0 & 0 & \dots  & 0 \\
    -1 & 0 & 1 & 0 & \dots  & 0 \\
    0 & -1 & 0 & 1 & \dots  & 0 \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & 0 & \dots  & -2 & 2
\end{pmatrix}\cdot\frac{1}{2dx},\;where\;dx = x_{i+1} - x_{i}$$
The first and last rows of the matrix represent the forward finite difference operator and backward finite difference operator, respectively, and every other row represents the central finite difference operator.

$f(x) = x^2$, $f'(x) = 2x$, $x\in[0,20]$

n = 100

```python
func = lambda x: x**2

# n = 100, xmax = 20
xmax = 20
n = 100
x = np.linspace(0, xmax, n, endpoint=True)

# Plotting f(x)
f = np.vectorize(func)
ac.plot(x, f(x), xlabel="x", ylabel="f(x)", title="f(x) = x^2")

# Plotting f'(x)
Df = ac.derivative(func, xmax=xmax, n=n)
ac.plot(x, Df, xlabel="x", ylabel="f'(x)", title="f'(x) = 2x")
```

### Comparison

We will now compare our method of computing the derivative with numpy's built in numpy.gradient function. We will be comparing the derivative of $x^2$

```python
# function and domain
func = lambda x: x**2
x = np.linspace(0, 99, 100, endpoint=True)
f = np.vectorize(func)
print("f(x) = x^2\n")

# our implementation
%timeit Df = ac.derivative(f, xmax=99, n=100)
print("Our f'(x) implementation: ")
print(Df)

print()

# numpy.gradient implementation
%timeit np.gradient(f(x))
print("numpy.gradient f'(x) implmentation: ")
print(np.gradient(f(x)))
```

#### Notes

As we can see both implementations produce the same results, but numpy's built in numpy.gradient function is faster than our implementation for computing the derivative.

### Squaring our derivative matrix

So, after everything we have seen, if we square our derivative matrix, we should get the second derivative. We will test this with $f(x) = x^2$

```python
# function and domain
func = lambda x: x**2
x = np.linspace(0, 99, 100, endpoint=True)
f = np.vectorize(func)
print("f(x) = x^2\n")

# our implementation
Df = ac.derivative(f, xmax=99, n=100, power=2) # squaring the matrix by passing in a power of 2
print("f''(x): ")
print(Df)
```

And as we can see, this seems to be the case, since $f''(x)$ in this case is $2$.
